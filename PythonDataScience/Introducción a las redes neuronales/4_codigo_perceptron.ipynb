```python
import numpy as np
import matplotlib.pyplot as plt
import random

class Perceptron:
    """
    Implementación detallada del Perceptrón Simple
    """
    def __init__(self, n_entradas, tasa_aprendizaje=0.1):
        """
        Inicializa el perceptrón con pesos aleatorios
        
        Parámetros:
        - n_entradas: número de características de entrada
        - tasa_aprendizaje: tasa de aprendizaje (η) para actualizar pesos
        """
        self.pesos = np.random.uniform(-1, 1, n_entradas)
        self.bias = np.random.uniform(-1, 1)
        self.tasa_aprendizaje = tasa_aprendizaje
        self.historia_pesos = []
        self.historia_errores = []
    
    def funcion_activacion(self, suma_ponderada):
        """
        Función escalón unitario (Heaviside)
        
        Retorna:
        1 si suma_ponderada > 0
        0 en caso contrario
        """
        return 1 if suma_ponderada > 0 else 0
    
    def predecir(self, entradas):
        """
        Realiza una predicción para un conjunto de entradas
        
        Parámetros:
        - entradas: array de características de entrada
        
        Retorna:
        La predicción (0 o 1)
        """
        suma_ponderada = np.dot(self.pesos, entradas) + self.bias
        return self.funcion_activacion(suma_ponderada)
    
    def entrenar(self, X, y, max_iteraciones=1000, verbose=True):
        """
        Entrena el perceptrón con un conjunto de datos
        
        Parámetros:
        - X: matriz de características de entrada
        - y: vector de etiquetas objetivo
        - max_iteraciones: número máximo de épocas de entrenamiento
        - verbose: si True, muestra información durante el entrenamiento
        """
        self.historia_pesos = [(self.pesos.copy(), self.bias)]
        self.historia_errores = []
        
        if verbose:
            print(f"\nIniciando entrenamiento:")
            print(f"Pesos iniciales: {self.pesos}")
            print(f"Bias inicial: {self.bias}")
            print("-" * 50)
        
        for epoca in range(max_iteraciones):
            error_total = 0
            cambios = 0
            
            # Iterar sobre cada ejemplo de entrenamiento
            for i in range(len(X)):
                # Realizar predicción
                prediccion = self.predecir(X[i])
                error = y[i] - prediccion
                error_total += abs(error)
                
                # Si hay error, actualizar pesos y bias
                if error != 0:
                    cambios += 1
                    # Guardar pesos antiguos para mostrar cambios
                    pesos_antiguos = self.pesos.copy()
                    bias_antiguo = self.bias
                    
                    # Actualizar pesos y bias
                    for j in range(len(self.pesos)):
                        delta = self.tasa_aprendizaje * error * X[i][j]
                        self.pesos[j] += delta
                        
                        if verbose and epoca < 3:
                            print(f"\nÉpoca {epoca + 1}, Ejemplo {i + 1}:")
                            print(f"  Entrada: {X[i]}")
                            print(f"  Objetivo: {y[i]}, Predicción: {prediccion}")
                            print(f"  Peso {j}: {pesos_antiguos[j]:.3f} → {self.pesos[j]:.3f}")
                            print(f"  Δw{j} = η * error * x{j} = {self.tasa_aprendizaje} * {error} * {X[i][j]} = {delta:.3f}")
                    
                    self.bias += self.tasa_aprendizaje * error
                    
                    if verbose and epoca < 3:
                        print(f"  Bias: {bias_antiguo:.3f} → {self.bias:.3f}")
            
            # Guardar historia de pesos y errores
            self.historia_pesos.append((self.pesos.copy(), self.bias))
            self.historia_errores.append(error_total)
            
            if verbose:
                print(f"\nÉpoca {epoca + 1}:")
                print(f"  Error total: {error_total}")
                print(f"  Cambios realizados: {cambios}")
            
            # Criterio de parada: sin errores
            if error_total == 0:
                if verbose:
                    print(f"\n¡Convergencia alcanzada en época {epoca + 1}!")
                break
                
            if epoca == max_iteraciones - 1:
                print("\n¡Máximo de épocas alcanzado!")
    
    def visualizar_frontera(self, X, y, titulo="Frontera de decisión"):
        """
        Visualiza la frontera de decisión del perceptrón
        """
        plt.figure(figsize=(10, 8))
        
        # Graficar puntos
        for i in range(len(X)):
            color = 'blue' if y[i] == 1 else 'red'
            marker = 'o' if y[i] == 1 else 'x'
            plt.scatter(X[i][0], X[i][1], c=color, marker=marker)
        
        # Graficar frontera de decisión
        if self.pesos[1] != 0:
            x = np.linspace(-0.5, 1.5, 100)
            y = (-self.pesos[0] * x - self.bias) / self.pesos[1]
            plt.plot(x, y, 'g--', label='Frontera de decisión')
        
        plt.title(titulo)
        plt.xlabel('x₁')
        plt.ylabel('x₂')
        plt.grid(True)
        plt.legend()
        plt.show()
    
    def visualizar_entrenamiento(self):
        """
        Visualiza la evolución del error durante el entrenamiento
        """
        plt.figure(figsize=(10, 6))
        plt.plot(self.historia_errores, 'b-')
        plt.title('Evolución del Error Durante el Entrenamiento')
        plt.xlabel('Época')
        plt.ylabel('Error Total')
        plt.grid(True)
        plt.show()

# Ejemplo de uso para AND y OR
if __name__ == "__main__":
    # Datos de entrenamiento
    X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    y_and = np.array([0, 0, 0, 1])
    y_or = np.array([0, 1, 1, 1])
    
    # Entrenar perceptrón para AND
    print("\nEntrenando perceptrón para compuerta AND:")
    p_and = Perceptron(2, tasa_aprendizaje=0.1)
    p_and.entrenar(X, y_and)
    p_and.visualizar_frontera(X, y_and, "Compuerta AND")
    p_and.visualizar_entrenamiento()
    
    # Entrenar perceptrón para OR
    print("\nEntrenando perceptrón para compuerta OR:")
    p_or = Perceptron(2, tasa_aprendizaje=0.1)
    p_or.entrenar(X, y_or)
    p_or.visualizar_frontera(X, y_or, "Compuerta OR")
    p_or.visualizar_entrenamiento()
    
    # Demostrar problema XOR
    y_xor = np.array([0, 1, 1, 0])
    print("\nIntentando entrenar perceptrón para compuerta XOR:")
    p_xor = Perceptron(2, tasa_aprendizaje=0.1)
    p_xor.entrenar(X, y_xor)
    p_xor.visualizar_frontera(X, y_xor, "Compuerta XOR (No separable linealmente)")
    p_xor.visualizar_entrenamiento()
```
