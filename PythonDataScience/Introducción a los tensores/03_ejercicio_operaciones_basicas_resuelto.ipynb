

### Resolución de los Ejercicios

#### 1. **Creación de Tensores**

- **Ejercicio 1**

   ```python
   tensor_zeros = torch.zeros(4, 3)
   tensor_ones = torch.ones(4, 3)
   print("Tensor de ceros:\n", tensor_zeros)
   print("Tensor de unos:\n", tensor_ones)
   ```

- **Ejercicio 2**

   ```python
   tensor_arange = torch.arange(20)
   tensor_linspace = torch.linspace(0, 1, 5)
   print("Tensor arange:\n", tensor_arange)
   print("Tensor linspace:\n", tensor_linspace)
   ```

#### 2. **Aritmética Básica**

- **Ejercicio 3**

   ```python
   tensor_a = torch.rand(3, 3)
   tensor_b = torch.rand(3, 3)
   suma = tensor_a + tensor_b
   resta = tensor_a - tensor_b
   print("Suma:\n", suma)
   print("Resta:\n", resta)
   ```

- **Ejercicio 4**

   ```python
   tensor_rand = torch.rand(5, 5)
   multiplicado = tensor_rand * 10
   dividido = multiplicado / 5
   print("Tensor multiplicado:\n", multiplicado)
   print("Tensor dividido:\n", dividido)
   ```

#### 3. **Reducciones Simples**

- **Ejercicio 5**

   ```python
   tensor_rand = torch.rand(2, 4)
   suma = torch.sum(tensor_rand)
   media = torch.mean(tensor_rand)
   print("Suma total:", suma)
   print("Media:", media)
   ```

- **Ejercicio 6**

   ```python
   tensor_random = torch.rand(3, 3)
   minimo_fila, _ = torch.min(tensor_random, dim=1)
   maximo_fila, _ = torch.max(tensor_random, dim=1)
   print("Mínimos por fila:", minimo_fila)
   print("Máximos por fila:", maximo_fila)
   ```

#### 4. **Comparaciones**

- **Ejercicio 7**

   ```python
   tensor_a = torch.rand(4, 4)
   tensor_b = torch.rand(4, 4)
   comparacion = tensor_a > tensor_b
   print("Comparación (mayor que):\n", comparacion)
   ```

- **Ejercicio 8**

   ```python
   tensor_rand = torch.rand(5)
   mayor_que_05 = tensor_rand >= 0.5
   print("Comparación con 0.5:\n", mayor_que_05)
   ```

#### 5. **Forma y Estructura**

- **Ejercicio 9**

   ```python
   tensor = torch.rand(3, 2, 4)
   elementos_totales = tensor.numel()
   forma = tensor.shape
   print("Número total de elementos:", elementos_totales)
   print("Forma del tensor:", forma)
   ```

- **Ejercicio 10**

   ```python
   tensor = torch.rand(6, 4)
   forma_original = tensor.size()
   tensor_reshaped = tensor.reshape(4, 6)
   print("Forma original:", forma_original)
   print("Tensor re-dimensionado:\n", tensor_reshaped)
   ```

#### 6. **Indexación y Slicing**

- **Ejercicio 11**

   ```python
   tensor = torch.arange(25).reshape(5, 5)
   submatriz = tensor[1:4, 1:4]
   print("Submatriz extraída:\n", submatriz)
   ```

- **Ejercicio 12**

   ```python
   tensor = torch.rand(3, 3)
   tensor[torch.arange(3), torch.arange(3)] = 1  # Diagonal principal
   print("Tensor con diagonal en unos:\n", tensor)
   ```
