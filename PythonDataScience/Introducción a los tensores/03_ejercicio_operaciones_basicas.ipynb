### Ejercicios por Función Básica

#### 1. **Creación de Tensores**
- **Ejercicio 1**: Crea un tensor de tamaño \(4 \times 3\) lleno de ceros y otro del mismo tamaño lleno de unos.
- **Ejercicio 2**: Genera un tensor que contenga los números del 0 al 19 usando `torch.arange()` y otro tensor con 5 valores equidistantes entre 0 y 1 usando `torch.linspace()`.

#### 2. **Aritmética Básica**
- **Ejercicio 3**: Crea dos tensores de tamaño \(3 \times 3\) con valores aleatorios y realiza la suma y resta entre ellos.
- **Ejercicio 4**: Genera un tensor aleatorio de tamaño \(5 \times 5\) y multiplícalo por un escalar de valor 10. Luego, divide el tensor resultante por 5.

#### 3. **Reducciones Simples**
- **Ejercicio 5**: Genera un tensor de valores aleatorios de tamaño \(2 \times 4\). Calcula la suma de todos los elementos y la media del tensor.
- **Ejercicio 6**: Crea un tensor de tamaño \(3 \times 3\) con valores aleatorios y calcula el valor mínimo y máximo de cada fila.

#### 4. **Comparaciones**
- **Ejercicio 7**: Crea dos tensores aleatorios de tamaño \(4 \times 4\). Verifica cuáles de los elementos del primer tensor son mayores que los del segundo tensor.
- **Ejercicio 8**: Genera un tensor aleatorio de tamaño \(5\) y verifica si cada elemento es mayor o igual a 0.5.

#### 5. **Forma y Estructura**
- **Ejercicio 9**: Genera un tensor de tamaño \(3 \times 2 \times 4\) y calcula su número de elementos y su forma.
- **Ejercicio 10**: Crea un tensor de tamaño \(6 \times 4\), obtén su tamaño y redimensionalo a \(4 \times 6\).

#### 6. **Indexación y Slicing**
- **Ejercicio 11**: Genera un tensor de tamaño \(5 \times 5\) con valores secuenciales y extrae la submatriz de los elementos de la segunda a la cuarta fila y columna.
- **Ejercicio 12**: Crea un tensor de tamaño \(3 \times 3\) y reemplaza los elementos de la diagonal principal por unos.

### Ejercicios Combinados

1. **Ejercicio 13**: Crea un tensor de tamaño \(3 \times 4\) con valores aleatorios, multiplícalo por 2 y calcula la suma de todos sus elementos.
2. **Ejercicio 14**: Genera un tensor aleatorio de tamaño \(4 \times 4\), calcula el promedio de cada fila, y compara estos promedios con un valor límite de 0.5.
3. **Ejercicio 15**: Crea un tensor de tamaño \(5\) con valores de 1 a 5. Usa aritmética para multiplicar cada elemento por 3 y luego calcula el número total de elementos.
4. **Ejercicio 16**: Genera dos tensores aleatorios de tamaño \(3 \times 3\), súmalos y verifica si cada elemento del resultado es mayor que 0.5.
5. **Ejercicio 17**: Crea un tensor de tamaño \(4 \times 4\) y cambia la forma del tensor a \(2 \times 8\). Después, calcula la media de todos sus elementos.
6. **Ejercicio 18**: Genera un tensor aleatorio de tamaño \(3 \times 4\), calcula el valor máximo de cada columna, y luego verifica si cada máximo es mayor a 0.8.
7. **Ejercicio 19**: Crea un tensor de tamaño \(5 \times 5\) con valores entre 0 y 1. Extrae los valores de la última columna y calcula su media.
8. **Ejercicio 20**: Genera un tensor aleatorio de tamaño \(2 \times 3 \times 4\), cambia su forma a \(4 \times 6\), y calcula la suma de todos sus elementos.
9. **Ejercicio 21**: Crea dos tensores de tamaño \(2 \times 2\), réstalos y encuentra el valor mínimo del tensor resultante.
10. **Ejercicio 22**: Genera un tensor aleatorio de tamaño \(3 \times 3\), multiplica cada elemento por 5 y calcula el número total de elementos.
